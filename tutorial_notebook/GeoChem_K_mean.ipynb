# import lib

```python
from matplotlib import pyplot as plt
import gdal
from sklearn import cluster
```

## Read the dataset

```python
img = gdal.Open("Playa_Image.tif")
data = img.ReadAsArray()
X = data.reshape((-1,1))
```

## Fit model and train

```python
k_means = cluster.KMeans(n_clusters=8)
k_means.fit(X)
X_cluster = k_means.labels_
X_cluster = X_cluster.reshape(data.shape)
```

## Visualization

```python
_, axs = plt.subplots(1, 7, figsize=(50, 50))
axs = axs.flatten()
for img1, ax in zip(data, axs ):
    ax.imshow(img1, cmap="hsv")
    # ax.imshow(img2)

plt.show()
```

```python
_, axs = plt.subplots(1, 7, figsize=(50, 50))
axs = axs.flatten()
for img1, ax in zip(X_cluster, axs ):
    ax.imshow(img1, cmap="hsv")
    # ax.imshow(img2)

plt.show()
```
